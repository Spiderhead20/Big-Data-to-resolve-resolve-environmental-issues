```python
#Importar modulos
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#Realizar regresi√≥n lineal
temperatureData = pd.read_csv('temperature-anomaly.csv')
#temperatureData
X = temperatureData.iloc[:, 2].values.reshape(-1, 1)
Y = temperatureData.iloc[:, 3].values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X, Y)
Y_pred = lr.predict(X)
```

```python
#Visualizar resultados
plt.scatter(X, Y)
plt.plot(X, Y_pred, color = 'red')
plt.show
```

```python
#Visualizar matriz de datos
sns.heatmap(temperatureData.corr())
```

```python
co2Data = pd.read_csv('co2-concentration-long-term.csv')
X2 = co2Data.iloc[:, 2].values.reshape(-1, 1)
Y2 = co2Data.iloc[:, 3].values.reshape(-1, 1)
lr2 = LinearRegression()
lr2.fit(X2, Y2)
Y_pred2 = lr2.predict(X2)
```

```python
plt.scatter(X2, Y2)
plt.plot(X2, Y_pred2, color = 'red')
plt.show
```

```python
sns.heatmap(co2Data.corr())
```

```python
climateData = pd.read_csv('future-greenhouse-gas-emission-scenarios.csv')
x3 = climateData.iloc[:, [3,4,5]].values
y3 = climateData.iloc[:, 12].values
climateData.head()
```

```python
sns.heatmap(climateData.corr())
```

```python
#Dividir los datos
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x3, y3, test_size = 0.2, random_state = 0)    
```

```python
#Regresion lineal multiple
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Prediccion y resultados
ypred = regressor.predict(x_test)
ypred
```

```python
#Calcular coeficientes e interceptores
print(regressor.coef_)
print(regressor.intercept_)
```

```python
#Calculando el valor de R al cuadrado
from sklearn.metrics import r2_score
r2_score(y_test, ypred)
```
